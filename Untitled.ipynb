```python
# importing the necessarly libraries
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.tree import DecisionTreeClassifier,plot_tree
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report
```

```python
#load the data
df = pd.read_csv('clothing_recommendation.csv')
```

```python
print("Sample Data:")
print(df.head())
```

```python
# Prepare Feature (X) and Target on (Y)
X = df[['Temperature', 'Humidity', 'Wind_Speed', 'Rain_Chance']]
Y = df['Clothing']
```

```python
#Split into training and testing data (80% train, 20% test)
X_train,X_test,Y_train,Y_test = train_test_split(X,Y, test_size = 0.2,random_state = 42)
```

```python
#Create and train the Decision Tree model
model = DecisionTreeClassifier()
model.fit(X_train,Y_train)
```

```python
# 6. Predict on test data
y_pred = model.predict(X_test)
```

```python
# 7. Print the classification report to check accuracy
print("\nModel Accuracy Report:")
print(classification_report(Y_test, y_pred))
```

```python
# 8. Test with your own custom input
test_data = [[12, 80, 15, 60]]
prediction = model.predict(test_data)
print("\nRecommended Clothing for input [12°C, 80% Humidity, 15 km/h Wind, 60% Rain]:", prediction[0])
```

```python
# 9. Visualize the Decision Tree
plt.figure(figsize=(16, 10))
plot_tree(model,
          feature_names=['Temperature', 'Humidity', 'Wind_Speed', 'Rain_Chance'],
          class_names=model.classes_,
          filled=True,
          rounded=True,
          fontsize=12)
plt.title("Decision Tree - Clothing Recommendation")
plt.show()
```
